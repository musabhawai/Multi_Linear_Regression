# Car Price with Dummy Variables

```python
df=pd.read_csv("carprices.csv")
df
```

```python
dummies=pd.get_dummies(df["Car Model"])
dummies
```

```python
merge1=pd.concat([df,dummies],axis="columns")
merge1
```

```python
final=merge1.drop('Car Model',axis="columns")
final
```

```python
X=final.drop("Sell Price($)",axis="columns")
X
```

```python
X=X.drop("Mercedez Benz C class",axis="columns")
X
```

```python
Y=df["Sell Price($)"]
Y
```

```python
reg=linear_model.LinearRegression()
reg.fit(X,Y)
```

```python
reg.predict([[45000,4,False,False]])
```

```python
reg.predict([[86000,7,False,True]])
```

```python
reg.score(X,Y)
```
